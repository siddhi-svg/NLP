```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stopwd_nltk = stopwords.words('english')
print(stopwd_nltk)
print("Number of stopwords:", len(stopwd_nltk))

text = """Florence Nightingale was a nurse who saved many lives in the 19th century.
She was named after the city of Florence in Italy, where her parents went
after they got married in 1818."""

words = [word for word in text.split() if word.lower() not in stopwd_nltk]
new_text = " ".join(words)

print(new_text)
print("Old length:", len(text))
print("New length:", len(new_text))

```
